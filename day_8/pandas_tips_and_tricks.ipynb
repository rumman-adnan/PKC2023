# 9- Convert Strings to Number

```python
df = pd.DataFrame({'A': [1,2,3, 10,11], 'B': ['4','5','6', '12','13'], 'C': [7,8,9,14,15]})
df
```

```python
df.dtypes
```

```python
# convert data type
df.astype({'B': 'int64'}).dtypes
# df['B'] = df.astype({'B': 'int64'}).dtypes
```

```python
df.dtypes
```

```python
# Another way, type casting
pd.to_numeric(df['B'], errors='coerce')
# If the error argument is passed as coerce , then invalid parsing will be set as NaN . If the error argument is passed as ignore , then invalid parsing will return the input.
```

```python
# change b type to int64
df['B'] = pd.to_numeric(df['B'], errors='coerce')
df.dtypes
```

# 16. Splitting a string into multiple columns

```python
# import libraries
import pandas as pd

df = pd.DataFrame({'name':['ahmed raza', 'Ali ahmed', 'Ali Tahir'],'location':['Karachi', 'Lahore', 'Islamabad'],'age':[20, 30, 40]})

df
```

```python
## Check what will happen if we split name column
# df.name.str.split(' ').head()
df.name.str.split(' ', expand=True)
```

```python
# It adds 2 new column
df[['first_name', 'last_name']] = df.name.str.split(' ', expand=True)
```

```python
df
```

```python
# Refine data manipulation
df = df[['first_name', 'last_name', 'location', 'age']]
```

# 23- Reshaping a dataframe

```python
distance = pd.DataFrame([['12345', 100, 200, 300], ['12346', 200, 300, 400], ['12347', 300, 400, 500]], 
                            columns=['id', 'Retail', 'factory', 'warehouse'])
distance
```

```python
distance.head().T
```

```python
distance2 = pd.DataFrame([[1, '12345', 'factory'], [2, '12346', 'warehouse'], [3, '12347', 'Retail']], 
                            columns=['id', 'id2', 'location_type'])
distance2.head()
```

```python
# By doing reshaping of data we can make good plots
distance_long = distance.melt(id_vars=['id'], var_name='location_type', value_name='distance')
```

```python
distance_long
```

```python
sns.barplot(x='id', y='distance', hue='location_type', data=distance_long)
```
