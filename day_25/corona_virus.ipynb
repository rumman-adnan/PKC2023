### Step 3: Cleaning the data
- Dealing with missing values, Correcting errors in the data, Dropping duplicates

```python
df.isnull().sum()
```

```python
print(df1.columns)
df1.LOCALITY_NAME.unique()
```

```python
df1.drop(columns=['LOCALITY_NAME', 'AL3_NAME', 'AL2_NAME'], inplace=True)
```

```python
df1.isnull().sum() / len(df1) *100
```

```python
df1.dtypes
```

```python
# Finding deaths in china
df1[df1['OUTCOME'] == 'DEATH'].count()
```

```python
df1.OUTCOME.unique()
```

```python
# Type casting the OUTCOME column
df1['OUTCOME'] = df1['OUTCOME'].astype('category')
```

```python
df1.OUTCOME.dtype
```

```python
plt.bar(df1.OUTCOME.unique(), df1.OUTCOME.value_counts())
plt.xlabel('Outcome', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title('Outcome of the patients', fontsize=20)
plt.show()
```

```python
# Pie chart
plt.pie('AL0_NAME', labels='VALUE', autopct='%1.1f%%')
```

```python
sns.boxplot(x='OUTCOME', y='VALUE', data=df1)
```

```python
# Select data for china and usa

df1[df1.AL0_NAME == 'United States' or df1.AL0_NAME == 'China']
```

```python
# Boxplot for China
sns.boxenplot(x='AL1_CODE', y='VALUE', data=df1[df1.AL0_NAME == 'China'])
plt.xlabel('Country', fontsize=15)
plt.ylabel('Value', fontsize=15)
plt.title('Value of the patients', fontsize=20)
plt.show()
```

```python
# IQR method to find outliers
Q1 = df1.VALUE.quantile(0.25)
Q3 = df1.VALUE.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```
