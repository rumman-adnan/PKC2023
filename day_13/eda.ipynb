# Exploratory Data Analysis
EDA is a process of analyzing data sets to summarize their main characteristics, often with visual methods. A statistical model can be used or not, but primarily EDA is for seeing what the data can tell us beyond the formal modeling or hypothesis testing task.. In python, we can use pandas, numpy, matplotlib, seaborn, and scikit-learn to do EDA.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# visualize and explore the data
sns.countplot(data=df, x='survived')
# no need for y label because it is a countplot, ther is count on y axis obviously
```

```python
sns.barplot(data=df, x="survived", y="age")
```

```python
sns.countplot(data=df, x="pclass")
```

```python
# data distribution
sns.histplot(data=df, x="age", bins=30)
```

```python
df['age'].hist(bins=30, color='darkred', alpha=0.7)
# alpha = opacity
```

```python
# make box plot of age and fare
# For 2 things, take categorical variable on x axis and numerical variable on y axis
sns.boxplot(data=df, y="age", x="survived", hue="who")
# sns.boxplot(data=df, x="age", y="fare")
# sns.boxplot(data=df, x="age")
```

```python
# now we have to explore outliers in boxplot. outliers are data points that are far away from the rest of the data points not in interquartile range
# we can see that there are outliers in age column
```

```python
sns.scatterplot(data=df, x="age", y="fare")
# These 2 points are outliers, destroy the plot
```

```python
sns.scatterplot(data=df, x="age", y="fare", hue="survived")
```

```python
sns.scatterplot(data=df, x="age", y="fare", hue="pclass")
# Understand why i put these variables on x and y axis and hue on pclass. what insights it can give us
plt.ylim(0, 300)
# We can see that data is not normally distributed
```

```python
# pairplot with respect to survived. survivied is in legends
sns.pairplot(data=df, hue="survived")
```

```python
# Stacked barchart/plots
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar')
```

```python
# If we make swarmplot on barplot, we can see the distribution of data, it swarm your data points
# It is a categorical plot, it tells us about the distribution of data
sns.swarmplot(data=df, x="pclass", y="fare")
```

```python
sns.swarmplot(data=df, x="survived", y="age")
```

```python
sns.swarmplot(data=df, x="survived", y="age", hue="who")
# It clearly shows woman survived more than men
```

```python
sns.swarmplot(data=df, x="pclass", y="age", hue="who")
```

```python

sns.catplot(data=df, x="pclass", y="fare", hue="embarked", kind="swarm")
```

```python
# point plot
sns.catplot(data=df, x="pclass", y="fare", hue="embarked", kind="point")
```

```python
# To check relations of 2 numerical variables we use lmplot
sns.lmplot(data=df, y="age", x="fare")

```

```python
# Apply condition on fare, include fare with less than 300
# sns.lmplot(data=df, x="age", y=df[df.fare<300], hue="survived", fit_reg=False)

titanic_df = df[df['fare'] < 300]
# Now you can use this filtered data to create an lmplot
sns.lmplot(y='age', x='fare', data=titanic_df)
```

```python
# How to add column to make age groups
df['age_group'] = pd.cut(df['age'], bins=[0, 10, 18, 39, 50, 80], labels=['child', 'teen', 'young_adult', 'adult', 'senior'])
```

```python
df.head()
```

```python
df.groupby("age_group")["fare"].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue="age_group", vars=["age", "fare", "pclass"])
```

```python
sns.countplot(data=df, x="age_group")
```
