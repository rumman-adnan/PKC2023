```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
from sklearn.cluster import KMeans # K-means clustering
from sklearn.preprocessing import StandardScaler # scaling data and standardizing data. Normalize data


# Importing the dataset in features.
X = df[['age', 'fare']]
X.age.fillna(value=X['age'].mean(), inplace=True)

X.age.dropna

#scaled data
scaler = StandardScaler() # create an object
X_scaled = scaler.fit_transform(X) # fit and transform the data

model = KMeans(n_clusters=3) # create an object
model.fit(X_scaled) # fit the model


# Assigning the cluster labels to the data

df['cluster'] = model.labels_
```

```python
joblib.dump(model, "my_model.pkl") # save
```

```python
model_loaded = joblib.load("my_model.pkl") # load
```
