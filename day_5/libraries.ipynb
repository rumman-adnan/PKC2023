```python
# plotting
sns.set_theme(style="darkgrid")

diamonds = sns.load_dataset("diamonds")
clarity_ranking = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

sns.boxenplot(x="clarity", y="price",color="b", data=diamonds, order=clarity_ranking, scale="linear")
```
